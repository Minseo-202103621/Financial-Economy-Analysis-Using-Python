```python
columns = ["조사연도","가구주_성별코드","노인가구여부","조손가구여부","한부모가구여부","다문화가구여부(2012년~)","가구주_만연령","가구주_교육정도_학력코드","가구주_혼인상태코드","경상소득(보완)","입주형태코드"]
```

```python
df1 = df[columns]
```

```python
df1
```

```python
df1 = df1.rename(columns = {
    "조사연도":'year',"가구주_성별코드":"sex","노인가구여부":"elders","조손가구여부":"grandparent","한부모가구여부":"one_parent","다문화가구여부(2012년~)":"global","가구주_만연령":"age","가구주_교육정도_학력코드":"education","가구주_혼인상태코드":"marriage","경상소득(보완)":"income","입주형태코드":"house"
})
```

```python
df1
```

```python
def edu(x):
    if x == 2:
        return 6
    elif x == 3:
        return 9
    elif x == 4:
        return 12
    elif x == 5:
        return 14
    elif x == 6:
        return 16
    elif x == 7:
        return 18
    elif x == 8:
        return 21
    else:
        return 0
df1['education'] = df1['education'].apply(edu)
```

```python
df2 = df1.loc[(df1["income"]> 0) & df1['education']>0]
df2
```

```python
import numpy as np
df2['income_log'] = np.log(df2['income'])
df2['education_log'] = np.log(df2['education'])
```

```python
df2
```

```python
sex_dummies = pd.get_dummies(df2['sex']).rename(columns=lambda x : 's' + str(x))
sex_dummies.tail()
```

```python
job_dummies = pd.get_dummies(df2['job']).rename(columns = lambda x : 'd' + str(x) )
job_dummies
```

```python
marriage_dummies = pd.get_dummies(df2['marriage']).rename(columns=lambda x : 'm' + str(x))
marriage_dummies.tail()
```

```python
elders_dummies = pd.get_dummies(df2['elders']).rename(columns = lambda x : 'e' + str(x[1]))
elders_dummies
```

```python
one_parent_dummies = pd.get_dummies(df2['one_parent']).rename(columns = lambda x :'o' + str(x[1]))
one_parent_dummies.tail()
```

```python
global_dummies = pd.get_dummies(df2['global']).rename(columns = lambda x : 'g'+ str(x[1]))
global_dummies.tail()
```

```python
df3 = pd.concat([df2,sex_dummies,marriage_dummies,elders_dummies,one_parent_dummies,global_dummies],axis=1)
df3.tail(5)
```

```python
df3['age2'] = df1['age'] ** 2
df3['globals'] = df3['global']
df3['education_log'] = df2['education_log']
df3['education_log']
```

```python
edu = df3[['education_log','age','age2','s2','m2','m3','m4','e2','o2','g2' ]]
X = edu.values # X : 교육연수
ones = np.ones(X.shape[0])
X_mat = np.column_stack((ones,X))
X_mat
```

```python
Y = df3['income'].values # Y : 소득
beta_hat = np.linalg.inv(X_mat.T.dot(X_mat)).dot(X_mat.T).dot(Y) # 243page  회귀계수 추정
beta_hat.round(4)
```

```python
Y_fitted = X_mat @ beta_hat  # Y 추정량
e = Y-Y_fitted   # 잔차
n = len(e)
k = beta_hat.shape[0]  # 설명변수 개수
sigma_hat2 = (e.T @ e) / (n-k) # 오차항의 분산에 대한 추정량
sigma_hat2.round(2)
```

```python
XtX_Inv = np.linalg.inv(X_mat.T@X_mat)
cov = sigma_hat2 * XtX_Inv
cov
```

```python
se = np.sqrt(cov)
se
```

```python
se1 = np.zeros(k)
se1[0,] = se[0,0]
se1[1,] = se[1,1]
se1
```

```python
lb = beta_hat - 2 * se1
hb = beta_hat + 2 * se1
bound = np.column_stack((lb,hb)) # 95% 유의수준에서 신뢰구간
bound.round(2)
```
