```python
temp['price'] = temp['price'].str.replace(',','').astype(int)

```

```python
temp['size'] = temp['size'].str.replace('.','.').astype(float)

```

```python
temp['size_pyung'] = temp['size']/3.33
```

```python
temp['size_pyung']
```

```python
temp["price_pyung"] = temp['price']/temp['size_pyung']
```

```python
temp.head()
```

```python
temp['month'] = temp['month'].str.replace(' ',' ').astype(int)
```

```python
temp['built_year'] = temp['built_year'].str.replace('','').astype(int)
```

```python
temp['apt_age'] = temp['month'] - temp['built_year']
```

```python
# 설명 변수 x = 아파트연령 , 종속변수 y =  평당가격
```

```python
df = temp[['apt_age','price_pyung','apt_name']]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
edu_model = ols("price_pyung~apt_age",data=df).fit()
print(edu_model.summary())
```

```python
## RMSE 계산
intercept = edu_model.params[0]
slope = edu_model.params[1]
df['fitted'] = intercept + df['apt_age'] * slope
df['sse'] = (df['price_pyung']-df['fitted'])**2
sse = df['sse']
n = len(df.index)
k=2
sigma_hat = np.sqrt(np.sum(sse)/(n-k)) # rmse
np.round(sigma_hat,2)
```

```python
## 회귀계수의 표준 오차
df['x_bar'] = df['apt_age'].mean()
x_bar = df['x_bar'].mean()
df['x_residual2'] = (df['apt_age']-df['x_bar'])**2
x_residual2 = np.sum(df['x_residual2'])
n = len(df.index)
se_a = np.sqrt(1/n + x_bar**2/x_residual2)*sigma_hat
np.round(se_a,2)
```

```python
se_b = sigma_hat/np.sqrt(x_residual2)
np.round(se_b,2)
```

```python
lb_a = intercept - 2*se_a
hb_a = intercept + 2*se_a
```

```python
np.round(lb_a,2)
```

```python
np.round(hb_a,2)
```

```python
lb_b = intercept - 2*se_b
hb_b = intercept + 2*se_b
```

```python
np.round(hb_b,2)
```

```python
np.round(lb_b,2)
```
